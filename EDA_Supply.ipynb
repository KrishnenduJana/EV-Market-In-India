```python
import matplotlib.pyplot as plt

# Generate the pie chart with an explosion effect
explode = [0.01] * len(df.groupby('Source').size())  # Explode each wedge by 10%
df.groupby('Source').size().plot(kind='pie', autopct='%.1f', explode=explode)

# Add labels and title
plt.ylabel('')  # Remove the default ylabel
plt.title('Distribution of Sources')

# Display the chart
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Example DataFrame
df = pd.DataFrame({'Brand': ['A', 'B', 'C', 'D', 'E', 'A'],
                   'Price': [25, 30, 20, 40, 50, 35]})

# Sort the DataFrame by 'Price' column in ascending order
max_price = df[df['Price'] == df['Price'].max()]
sorted_df = max_price.sort_values('Price')

# Plotting the barplot
sns.barplot(data=sorted_df, x="Brand", y="Price")
plt.xlabel('Brand')
plt.ylabel('Price')
plt.title('Barplot (Ascending Order)')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Example DataFrame
df = pd.DataFrame({'Brand': ['A', 'B', 'C', 'D', 'E', 'A', 'B', 'C', 'D', 'E'],
                   'Price': [10, 30, 20, 40, 50, 15, 35, 25, 45, 55]})

# Filter the DataFrame to include only rows with maximum price values
max_price = df[df['Price'] == df['Price'].max()]

# Sort the max_price DataFrame in ascending order for each brand
sorted_df = max_price.groupby('Brand').apply(lambda x: x.sort_values('Price'))

# Plotting the barplot with sorted maximum price values for each brand
sns.barplot(data=sorted_df, x="Brand", y="Price")
plt.xlabel('Brand')
plt.ylabel('Price')
plt.title('Barplot (Maximum Price - Sorted)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
df = pd.DataFrame({
    'ModelName': ['Model A', 'Model B', 'Model C', 'Model D', 'Model E'],
    'Brand': ['Brand X', 'Brand Y', 'Brand X', 'Brand Z', 'Brand Y'],
    'Price': [10000, 20000, 15000, 18000, 25000],
    'Power(kwh)': [200, 180, 220, 190, 210],
    'Mileage(km)': [300, 350, 320, 280, 330]
})

# Plotting histogram for Price in ascending order
sorted_price = df['Price'].sort_values()
plt.hist(sorted_price, bins='auto', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram for Price (Ascending Order)')
plt.show()

# Plotting bar plot for Brand in ascending order
sorted_brand = df['Brand'].value_counts().sort_index()
plt.bar(sorted_brand.index, sorted_brand)
plt.xlabel('Brand')
plt.ylabel('Count')
plt.title('Bar Plot for Brand (Ascending Order)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
df = pd.DataFrame({
    'ModelName': ['Model A', 'Model B', 'Model C', 'Model D', 'Model E', 'Model F'],
    'Brand': ['Brand X', 'Brand Y', 'Brand X', 'Brand Z', 'Brand Y', 'Brand Z'],
    'Price': [10000, 20000, 15000, 18000, 25000, 22000],
    'Power(kwh)': [200, 180, 220, 190, 210, 195],
    'Mileage(km)': [300, 350, 320, 280, 330, 310]
})

# Calculate maximum price for each brand
max_prices = df.groupby('Brand')['Price'].max()

# Sort maximum prices in ascending order
sorted_prices = max_prices.sort_values()

# Plotting histogram for maximum prices in ascending order
plt.hist(sorted_prices, bins='auto', edgecolor='black')
plt.xlabel('Maximum Price')
plt.ylabel('Frequency')
plt.title('Histogram of Maximum Price for Each Brand (Ascending Order)')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
df = pd.DataFrame({
    'ModelName': ['Model A', 'Model B', 'Model C', 'Model D', 'Model E', 'Model F'],
    'Brand': ['Brand X', 'Brand Y', 'Brand X', 'Brand Z', 'Brand Y', 'Brand Z'],
    'Price': [10000, 20000, 15000, 18000, 25000, 22000],
    'Power(kwh)': [200, 180, 220, 190, 210, 195],
    'Mileage(km)': [300, 350, 320, 280, 330, 310]
})

# Calculate mean price for each brand
mean_prices = df.groupby('Brand')['Price'].mean().reset_index()

# Sort the mean prices in ascending order
sorted_mean_prices = mean_prices.sort_values('Price')

# Plotting barplot for mean prices in sorted order
sns.barplot(data=sorted_mean_prices, x='Brand', y='Price')
plt.xlabel('Brand')
plt.ylabel('Mean Price')
plt.title('Mean Price for Each Brand (Sorted)')
plt.show()
```
